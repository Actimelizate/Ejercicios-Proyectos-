### SOLUCION CON MAP

```
lista_medidas_cadena = lineas_medida.map(lambda elemento: elemento.split(','))

lista_medidas_cadena.take(2)
```

#
##### PRIMER ELEMENTO DEL RDD DEL QUE PARTO
##### ['21.5',  '29.6',  '19.3',  '22.7',  '37.5',  '15.',  '27.6',  '33.8',  '12.5',  '17.7',  '36.',  '14.8',  '25.1',  '17.3',
#####  '16.2',  '33.2',  '30.4',  '18.4',  '17.3',  '18.6',  '37.2',  '39.7']

```

def calculo_totales_linea(lista):
    # parametro entrada lista es cada elemento (una lista) del RDD al que hemos llegado, 
    # que se corresponde con una linea o renglon del archivo original
    
    suma_lista = 0
    num_medidas_lista = 0

    # recorro la lista de medidas
    for medida in lista:
        # convierto a numero
        cambio = float(medida)
        
        # añado a la suma por linea
        suma_lista = suma_lista + cambio
        
        # cuento una medida más
        num_medidas_lista = num_medidas_lista + 1
        
        
    totales_linea = [suma_lista, num_medidas_lista]
        
    return totales_linea

```

```
totales_linea = lista_medidas_cadena.map(calculo_totales_linea)

totales_linea.take(5)
```

```
suma_total_medidas, num_medidas_total = totales_linea.reduce(lambda e1, e2: [e1[0]+e2[0], e1[1]+e2[1]])
```

```
print('La media es: ', suma_total_medidas/num_medidas_total)
```

### SOLUCION CON FLATMAP

```
lista_medidas_cadena2 = lineas_medida.flatMap(lambda elemento: elemento.split(','))

lista_medidas_cadena2.take(10)
```

```
medidas_numero = lista_medidas_cadena2.map(lambda elemento: float(elemento))

medidas_numero.take(10)

```

```
suma_total = medidas_numero.reduce(lambda m1, m2: m1 + m2)

numero_medidas = medidas_numero.count()

print('La media es: ', suma_total/numero_medidas)
```
